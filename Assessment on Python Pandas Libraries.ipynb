# 1)Pull data in Python using Pandas

```python
import pandas as pd
df = pd.read_csv('project1.csv')
print(df)
```

```python
df['COGS'] = df['COGS'].fillna(df['COGS'].mean())
```

```python
df.fillna(({'Segment':'Government','Discount Band':'Medium'}),inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
import datetime

date_list = []
x = df['Date']
dates = pd.to_datetime(df['Date']).dt.date
for date in range(len(x)):
    date_list.append(dates[date] + datetime.timedelta(hours=int(x[date])))

date_list
```

```python
df['Date'] = pd.Series(date_list)
df
```

# 3)Making relevant Bar chart with its elements

```python
line1=df.groupby('Country').agg({' Sales':["sum"]})
pd.DataFrame(line1)
```

```python
country_list = line1.index.tolist()
sales_list = line1.iloc[:,0].tolist()
```

```python
x=np.arange(len(country_list))
y=np.arange(len(sales_list))

plt.figure(figsize=(18,10))
plt.title('COUNTRYWISE SALES',fontsize=32,fontweight='bold')
plt.xlabel('Countries',fontweight="bold",fontsize=30)
plt.ylabel("NET SALES IN ($)",fontweight="bold",fontsize=30)

plt.xticks(x -0.2,country_list,fontsize =20)
plt.yticks(fontsize =25)
plot1 = plt.bar(x - 0.2, sales_list, 0.4,color = '#ff0000', label = 'NET SALES')

for i in range(len(sales_list)):
    plt.text(i, sales_list[i], sales_list[i], ha ='right', va ='bottom')
plt.legend(fontsize = 15)
plt.show
```

```python
line2=df.groupby('Country').agg({'Profit':['sum']})
pd.DataFrame(line2)
```

```python
country_list=line1.index.tolist()
profit_list=line1.iloc[:,0].tolist()
```

```python
x = np.arange(len(country_list))
y = np.arange(len(profit_list))

plt.figure(figsize=(18,10))
plt.title('COUNTRYWISE PROFIT',fontsize=32,fontweight='bold')
plt.xlabel('Countries',fontsize=30,fontweight='bold')
plt.ylabel('profit in ($)',fontsize=30,fontweight='bold')

plt.xticks(x -0.2,country_list,fontsize=20)
plt.yticks(fontsize=25)
plot2 = plt.bar(x -0.2, profit_list,0.4,color='orange',label='profit')

for i in range(len(sales_list)):
    plt.text(i,profit_list[i], profit_list[i],ha='right', va='bottom')
    
plt.legend(fontsize=15)
plt.show()
```

```python
country_list=line1.index.tolist()
profit_list=line2.iloc[:,0].tolist()
sales_list=line1.iloc[:,0].tolist()
```

```python
x_axis=np.arange(len(country_list))
y_axis=np.arange(len(sales_list))

plt.figure(figsize=(20,10))
plt.title('COUNTRIEWISE SALES & PROFIT',fontsize=35,fontweight='bold')
plt.xlabel('Countries',fontweight='bold',fontsize=30)
plt.ylabel('($)',fontweight='bold',fontsize=30)

plt.xticks(x_axis,country_list,fontsize =20)
plt.yticks(fontsize =25)
plot1 = plt.bar(x_axis - 0.2, sales_list, 0.4, color = '#7a49a5', label = 'Net Sales')
plot2 = plt.bar(x_axis + 0.2, profit_list, 0.4,color = 'Orange', label = 'Profits ')

for i in range(len(sales_list)):
    plt.text(i, sales_list[i], sales_list[i], ha ='right', va ='bottom')
for i in range(len(sales_list)):
    plt.text(i, profit_list[i], profit_list[i], ha ='left', va ='bottom')
plt.legend(fontsize = 15)
plt.show()
```

# 5)Making relevant Boxplot chart with its elements

```python
fig = plt.figure(figsize=(15,10))
set_1 = df['Gross Sales']
set_2 = df[' Sales']
set_3 = df['Profit']

set = [set_1,set_2,set_3]

red_circle = dict(markerfacecolor = '#00a2ff',marker='o',markeredgecolor='orange')
boxplot = plt.boxplot(set,patch_artist=True,flierprops = red_circle,vert=True,notch=True,showmeans=True)
plt.xticks([1,2,3],['Gross Sales','Sales','Profit'],fontsize=15)
plt.ylabel('$')

colors = ['green','blue','red']

for patch,color in zip(boxplot['boxes'], colors):
    patch.set_facecolor(color)
    
plt.show()


```
