# Binning

```python
s0 = df['Speed (IAS) in knots'].quantile(0)
s25 = df['Speed (IAS) in knots'].quantile(0.25)
s50 = df['Speed (IAS) in knots'].quantile(0.5)
s75 = df['Speed (IAS) in knots'].quantile(0.75)
s100 = df['Speed (IAS) in knots'].quantile(1)

df['Speed (IAS) in knots bin']=pd.cut(df['Speed (IAS) in knots'],bins=[s0,s25,s50,s75,s100],labels=['Low','High'],duplicates='drop')
```

```python
df
```

```python
s0 = df['Cost: Total $'].quantile(0)
s25 = df['Cost: Total $'].quantile(0.25)
s50 = df['Cost: Total $'].quantile(0.5)
s75 = df['Cost: Total $'].quantile(0.75)
s100 = df['Cost: Total $'].quantile(1)

df['Cost: Total $ bin']=pd.cut(df['Cost: Total $'],bins=[s0,s25,s50,s75,s100],labels=['High'],duplicates='drop')
```
